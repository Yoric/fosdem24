```python
from qadence import Register
import networkx as nx
graph = nx.complete_graph(3)
# Set custom node positions, in this case a simple line:
for i, node in enumerate(graph.nodes):
    graph.nodes[node]["pos"] = (1.0 * i, 0.0)
reg = Register(graph)
```

```python
reg.draw()
```
